```python
## One of the Reviewer asked about Emojies and hashtags. This scrips tries to find as many as possible.
import emoji, csv, re
import emoji_data_python
```

```python
DATASET = '../../Dataset/Text-Data/RP-Mod-Crowd.csv'
NAME = DATASET.split('/')[-1]
```

```python
## Callects all text emojies from emoji_data_python
text_emojies = []
for emj in emoji_data_python.emoji_data:
    if hasattr(emj, 'texts') and emj.texts is not None:
        text_emojies.extend(emj.texts)
text_emojies = set(text_emojies)
```

```python
emojies_paths = []
hash_tags = []
lines = 0
## Iterate over Dataset row-wise
with open(DATASET) as f:
    reader = csv.DictReader(f)
    for row in reader:
        row = {k.lower():v for k,v in row.items()}
        if 'text' not in row:
            continue
        lines += 1
        ## Split lines into words
        word_set = set(row['text'].split(' '))
        ## Select and add ascii emojis and normal ones 
        num_emojies = len(emoji.emoji_lis(row['text'],  language='de')) + \
                        len(set.intersection(word_set, text_emojies))
        ## Count Hasttags, but substract all double ## as these are no hastags
        num_hash_tags = len(re.findall('(?i) \#\w+', row['text'])) - len(re.findall('(?i) \#{2,}\w+', row['text']))
        if num_emojies >= 1:
            print(row['text'])
            emojies_paths.append(row)

        if num_hash_tags >= 1:
            print(row['text'])
            hash_tags.append(row)
            
    print(f'### Nunber of Rows: {lines}')
```
