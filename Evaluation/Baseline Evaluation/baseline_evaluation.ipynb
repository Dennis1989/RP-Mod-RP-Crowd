```python
import numpy as np
import pickle
import sklearn
from sklearn.metrics import accuracy_score
import pandas as pd

out = "experiments"
results = "../Repeated-Baseline-Results/"
## load data
```

```python
df = pd.read_csv("../../Dataset/Text-Data/RP-Crowd-2-folds.csv")

## load labels and cast to int
y_dat = df["label"].values
y_dat = y_dat.astype(int)

## test data filter
filter_q =  df['ten_folds'] < 8 
test_y = y_dat[-filter_q]


## tfidf
min2_tf_logreg = [accuracy_score(test_y ,pickle.load( open( results + "logreg_abusive_min_2_tfidf_"+str(i)+".p", "rb" ) )["decision"]) for i in range(0,5)]
print("min2_tf_logreg")
print("mean: " + str(np.mean(min2_tf_logreg)))
print("std: "+str(np.std(min2_tf_logreg)))
print("------------------------------")

min2_tf_nb = [accuracy_score(test_y ,pickle.load( open( results + "nb_abusive_min_2_tfidf_"+str(i)+".p", "rb" ) )["decision"]) for i in range(0,5)]
print("min2_tf_nb")
print("mean: " + str(np.mean(min2_tf_nb)))
print("std: "+str(np.std(min2_tf_nb)))
print("------------------------------")

min2_tf_xg = [accuracy_score(test_y ,pickle.load( open( results + "xg_abusive_min_2_tfidf_"+str(i)+".p", "rb" ) )["decision"]) for i in range(0,5)]
print("min2_tf_xg")
print("mean: " + str(np.mean(min2_tf_xg)))
print("std: "+str(np.std(min2_tf_xg)))
print("------------------------------")

## embedding
min2_embed_logreg = [accuracy_score(test_y ,pickle.load( open( results + "logreg_abusive_min_2_embed_"+str(i)+".p", "rb" ) )["decision"]) for i in range(0,5)]
print("min2_embed_logreg")
print("mean: " + str(np.mean(min2_embed_logreg)))
print("std: "+str(np.std(min2_embed_logreg)))
print("------------------------------")

min2_embed_nb = [accuracy_score(test_y ,pickle.load( open( results + "nb_abusive_min_2_embed_"+str(i)+".p", "rb" ) )["decision"]) for i in range(0,5)]
print("min2_embed_nb")
print("mean: " + str(np.mean(min2_embed_nb)))
print("std: "+str(np.std(min2_embed_nb)))
print("------------------------------")

min2_embed_xg = [accuracy_score(test_y ,pickle.load( open( results + "xg_abusive_min_2_embed_"+str(i)+".p", "rb" ) )["decision"]) for i in range(0,5)]
print("min2_embed_xg")
print("mean: " + str(np.mean(min2_embed_xg)))
print("std: "+str(np.std(min2_embed_xg)))
print("------------------------------")



np.std(min2_tf_xg)

```
